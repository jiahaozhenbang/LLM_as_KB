```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np

result_csv = '../output/results_icl.csv'
df = pd.read_csv(result_csv)
```

```python
x = df[df['ground-truth'] == 0]['zero-shot-p_neg']
y = df[df['ground-truth'] == 0]['zero-shot-p_pos']
plt.scatter(x, y, s=1, label='label=0',c="blue")
x = df[df['ground-truth'] == 1]['zero-shot-p_neg']
y = df[df['ground-truth'] == 1]['zero-shot-p_pos']
plt.scatter(x, y, s=1, label='label=1',c="green")

x = np.arange(0,max(max(df['zero-shot-p_neg']), max(df['zero-shot-p_pos'])),0.0001)
plt.plot(x, x, label="y=x", c='red')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np


# unused but required import for doing 3d projections with matplotlib < 3.2
import mpl_toolkits.mplot3d  # noqa: F401

from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
result_csv = '../output/results_zero-shot_dbpedia.csv'
df = pd.read_csv(result_csv)


NUM_LABEL=14

x = np.array([[row[f'zero-shot-p_{i}'] for i in range(NUM_LABEL)] for index, row in df.iterrows()])
y = np.array(df['ground-truth'])



# To getter a better understanding of interaction of the dimensions
# plot the first three PCA dimensions
fig = plt.figure(1, figsize=(8, 6))
ax = fig.add_subplot()
# ax = fig.add_subplot(111, projection="3d", elev=-150, azim=110)

id2verb = ['company', 'school', 'artist', 'athlete', 'politics',
                        'transportation', 'building', 'nature', 'village', 'animal',
                        'plant', 'album', 'film', 'book']
X_reduced = TSNE(n_components=2).fit_transform(x)
print(X_reduced.shape, type(X_reduced))
for i in range(NUM_LABEL):

    ax.scatter(
        X_reduced[y==i, 0],
        X_reduced[y==i, 1],
        c=[plt.cm.Set1(i)] * sum(y==i),
        label=id2verb[i],
        edgecolor="k",
        s=40,
    )

ax.set_title("First three PCA directions")
ax.set_xlabel("1st eigenvector")
ax.xaxis.set_ticklabels([])
ax.set_ylabel("2nd eigenvector")
ax.yaxis.set_ticklabels([])

plt.legend()
plt.show()
```
