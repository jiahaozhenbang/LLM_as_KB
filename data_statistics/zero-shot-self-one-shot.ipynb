```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np

result_csv = '../output/results_icl.csv'
df = pd.read_csv(result_csv)
```

```python
x = df[df['ground-truth'] == 0]['zero-shot-p_neg']
y = df[df['ground-truth'] == 0]['zero-shot-p_pos']
plt.scatter(x, y, s=1, label='label=0',c="blue")
x = df[df['ground-truth'] == 1]['zero-shot-p_neg']
y = df[df['ground-truth'] == 1]['zero-shot-p_pos']
plt.scatter(x, y, s=1, label='label=1',c="green")

x = np.arange(0,max(max(df['zero-shot-p_neg']), max(df['zero-shot-p_pos'])),0.0001)
plt.plot(x, x, label="y=x", c='red')
plt.legend()
plt.show()
```

```python
x = df[df['ground-truth'] == 0]['1st-self-one-shot-p_neg']
y = df[df['ground-truth'] == 0]['1st-self-one-shot-p_pos']
plt.scatter(x, y, s=1, label='label=0',c="blue")
x = df[df['ground-truth'] == 1]['1st-self-one-shot-p_neg']
y = df[df['ground-truth'] == 1]['1st-self-one-shot-p_pos']
plt.scatter(x, y, s=1, label='label=1',c="green")

x = np.arange(0,max(max(df['1st-self-one-shot-p_neg']), max(df['1st-self-one-shot-p_pos'])),0.0001)
plt.plot(x, x, label="y=x", c='red')
plt.legend()
plt.show()
```

```python
x = df[df['ground-truth'] == 0]['2nd-self-one-shot-p_neg']
y = df[df['ground-truth'] == 0]['2nd-self-one-shot-p_pos']
plt.scatter(x, y, s=1, label='label=0',c="blue")
x = df[df['ground-truth'] == 1]['2nd-self-one-shot-p_neg']
y = df[df['ground-truth'] == 1]['2nd-self-one-shot-p_pos']
plt.scatter(x, y, s=1, label='label=1',c="green")

x = np.arange(0,max(max(df['2nd-self-one-shot-p_neg']), max(df['2nd-self-one-shot-p_pos'])),0.0001)
plt.plot(x, x, label="y=x", c='red')
plt.legend()
plt.show()
```

### avg probs

```python
x = df[df['ground-truth'] == 0]['1st-self-one-shot-p_neg'] + df[df['ground-truth'] == 0]['2nd-self-one-shot-p_neg']
y = df[df['ground-truth'] == 0]['1st-self-one-shot-p_pos'] + df[df['ground-truth'] == 0]['2nd-self-one-shot-p_pos']
plt.scatter(x, y, s=1, label='label=0',c="blue")
cur_max = max(max(x), max(y))
x = df[df['ground-truth'] == 1]['1st-self-one-shot-p_neg'] + df[df['ground-truth'] == 1]['2nd-self-one-shot-p_neg']
y = df[df['ground-truth'] == 1]['1st-self-one-shot-p_pos'] + df[df['ground-truth'] == 1]['2nd-self-one-shot-p_pos']
plt.scatter(x, y, s=1, label='label=1',c="green")
cur_max = max(cur_max, max(max(x), max(y)))

x = np.arange(0,cur_max,0.0001)
plt.plot(x, x, label="y=x", c='red')
plt.legend()
plt.show()
```
