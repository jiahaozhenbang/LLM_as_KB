```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
def competence_function(t, p=1, T=1000, c0=0.01):
    tmp = pow((1- pow(c0, p)) /T * t + pow(c0, p), 1/p)
    return min(1, tmp)

from functools import partial
x = np.arange(0, 1200, 1)
y1 = list(map(partial(competence_function, p=1), x))
y2 = list(map(partial(competence_function, p=2), x))
y4 = list(map(partial(competence_function, p=4), x))
y8 = list(map(partial(competence_function, p=8), x))
y16 = list(map(partial(competence_function, p=16), x))
x0 = [1000] * 100
y0= np.arange(0, 1, 0.01)
plt.plot(x0,y0,linestyle = "--", color='r')
plt.plot(x,y1,label="p=1")
plt.plot(x,y2,label="p=2")
plt.plot(x,y4,label="p=4")
plt.plot(x,y8,label="p=8")
plt.plot(x,y16,label="p=16")
plt.xlabel("Time")
# plt.ylabel("Competence")
# plt.title('sin & cos')
plt.legend()   #打上标签
plt.show()
```
